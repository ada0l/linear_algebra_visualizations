# Matrix multiplication 

```python
import numpy as np

vec_1 = np.matrix([[0, 1], [1, 0]])
vec_2 = np.matrix([[2], [1]])

print(vec_1 * vec_2)
```

# Matrix multiplication as vector transposition in 2D 

```python
import numpy as np
from matplotlib import pyplot as plt

ax = plt.axes()

start = (0, 0)

# rotate 90 degrees to the right 
n = np.matrix([
    [0, -1],
    [1,  0]
])

m = np.matrix([
    [1],
    [1]
])

matrix_for_draw = [
    ('blue', n),
    ('green', m),
    ('red', n * m)
]

for mat in matrix_for_draw:
    for row in np.transpose(np.array(mat[1])):
        ax.arrow(*start, *row, ec = mat[0])

plt.scatter(*start)
plt.show()
```

# Matrix multiplication as vector transposition in 2D 

```python
import numpy as np
from matplotlib import pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

start = (0, 0, 0)


ox = np.matrix([
    [1],
    [0],
    [0]
])

oy = np.matrix([
    [0],
    [1],
    [0]
])

oz = np.matrix([
    [0],
    [0],
    [1]
])

trans = np.matrix([
    [0.5, 0, 0],
    [0.5, 0.5, 0],
    [0, 0, 0.5]
])

a = np.matrix([
    [1],
    [1],
    [1]
])

matrix_for_draw = [
    #('green', ox),
    #('red', oy),
    #('blue', oz),
    ('red', a),
    ('blue', trans),
    ('pink', trans * a)
]

for mat in matrix_for_draw:
    for row in np.transpose(np.array(mat[1])):
        ax.quiver(*[[0] for _ in range(len(row))],
                  *row, color=mat[0])


        
plt.scatter(*start)

ax.view_init(300, 35)
plt.draw()
```
